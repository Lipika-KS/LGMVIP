Import Libraries

```python
from google.colab.patches import cv2_imshow
import cv2
```

Reading the Image

```python
img = cv2.imread('/content/minholly.png')
cv2_imshow(img)
```

Apply GrayScale

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)   
cv2_imshow(img_gray)
```

Invert Image

```python
img_invert = cv2.bitwise_not(img_gray)  
cv2_imshow(img_invert)
```

Blurring the Image by applying Gaussian filter to the inverted image

```python
img_smoothing = cv2.GaussianBlur(img_invert, (133, 133),sigmaX=0, sigmaY=0)  
cv2_imshow(img_smoothing)
```

```python
def dodgeV2(x, y):                                  # dodging and merging
    return cv2.divide(x, 255 - y, scale=256)
```

Merge the grayScale image and Blur Image.

```python
# getting our sketch
final_img = dodgeV2(img_gray, img_smoothing)
cv2_imshow(img)
cv2_imshow(final_img)
```
